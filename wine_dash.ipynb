# Визуализация данных об отзывах о вине

## 1. Предобработка табличных данных

```python
# Импорты
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
# загружаем данные из файла

df = pd.read_csv('winemag-data_first150k.csv.gz')
print(
    df.shape,
    df.columns,
    sep='\n')
df.head(1)
```

## Одномерные графики 

* категориальные шкалы измерения
* количественные шкалы измерения

```python
# Столбчатая диаграмма

# df['province'].value_counts().plot.bar()
# df['province'].value_counts().head(10).plot.bar()
# df['province'].value_counts().head(10).plot.barh()
(df['province'].value_counts().head(10) / len(df) * 100).plot.barh()
```

```python
# Круговая диаграмма 
df['province'].value_counts().head(10).plot.pie()
```

```python
# категоризация количественных относительных переменных

# df['points'].value_counts().plot.bar()
df['points'].value_counts().sort_index().plot.bar()
```

```python
# Гистограмма
df['points'].plot.hist()
```

```python
df[df['province'] == 'Northern Spain']['points'].value_counts().sort_index().plot.line(color = 'green')
df[df['province'] == 'Northeastern Italy']['points'].value_counts().sort_index().plot.line(color = 'red')
```

```python
df['points'].plot.box()
```

```python
df['points'].describe()
df['points'].plot.hist()
```

## Многомерные графики

```python
# Диаграмма рассеяния
# df.plot.scatter(x='points', y='price')
df[df['price'] < 100].sample(500).plot.scatter(x='points', y='price')
```

```python
# Гексагон

df[df['price'] < 100].sample(500).plot.hexbin(x='points', y='price', gridsize=15)
```
